```python
import pandas as pd 
import numpy as np 
```

```python
df = pd.read_csv('test.csv') #leemos el archivo
df.head() #Mostramos primeras 5 filas 
```

```python
df.isnull().sum() #Buscamos nulos 
```

```python
# Fill missing values in 'Age' with the mean
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
# Fill missing values in 'Embarked' with the mode (most frequent value)
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
# Fill missing 'Fare' values with the mean (if any exist)
if 'Fare' in df.columns:
    df['Fare'] = df['Fare'].fillna(df['Fare'].mean())
```

```python
# Strip extra spaces from column names
df.columns = df.columns.str.strip()
```

```python
# Strip extra spaces from 'Name' column
df['Name'] = df['Name'].str.strip()
```

```python
#Standardize 'Sex' column to lowercase
df['Sex'] = df['Sex'].str.lower()
```

```python
# View the first few rows to confirm changes
df.head()
```

```python
# Export the cleaned dataset to a new CSV file
df.to_csv('test_clean.csv', index=False)

```
